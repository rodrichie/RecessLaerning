```python
import pandas as pd
import numpy as np
```

```python
#load data set
df = pd.read_csv('D:\\learn python\\UceResultsBySchool2011-2016.csv')
df.head(20)
```

```python
#preparation and preprocessing or data cleaning
df.columns
```

```python
#removing white spaces
df.columns = df.columns.to_series().apply(lambda x:x.strip())
df.columns
```

```python
#replac column space with underscores
df.columns = df.columns.str.replace(' ','_')
df.columns
```

```python
#check for duplicates
for isduplicate in df.duplicated():

    if isduplicate :
        print('Duplicates found and dropped')
        df.drop_duplicates(inplace= True)

        break
    else :
        pass
print('no duplicates found')
df[df.duplicated()==True]    
```

```python
#check for null and Nan values in the dataet
df.isna().any()
```

```python
#check for number of rows that are null
null_rows = df[df.isna().any(axis=1)]
before_compution=len(null_rows.axes[0])

print('number ofrows that are null values :' +str(before_compution))
      
```

```python
#remove rows without school names
df.dropna(thresh=None, subset=['SCHOOL'], inplace=True)
```
