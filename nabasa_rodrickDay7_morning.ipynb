```python
#convert columns to numeric type
numeric_columns =['Age','YOB', 'CGPA', 'Duration', 'Pulse', 'Maxpulse', 'Calories']
df4[numeric_columns]= df4[numeric_columns].apply(pd.to_numeric, errors='coerce')
```

```python
# Fill missing values using the mean of each column
df4_mean_filled = df4.copy()  # Create a copy of the original DataFrame
numeric_columns = ['Age', 'YOB', 'CGPA', 'Duration', 'Pulse', 'Maxpulse', 'Calories']

for column in numeric_columns:
    column_mean = df4[column].mean()
    df4_mean_filled[column].fillna(column_mean, inplace=True)

# Display the DataFrame without truncation
with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    display(df4_mean_filled)
```
